```python
#Making a Data Frame of the pokemon list
pokedata = pd.DataFrame(pokedex["pokemon"])
pokedata.head(5) #checking out the first five observations
```

```python
#Breaking out types - since some have two types, need to break into two columns
pokedata[["type1","type2"]] = pd.DataFrame(pokedata.type.tolist(), index = pokedata.index) #This is saying make and add two columns - using the type data, turn it into a list. The index keeps the rows intact
pokedata.head(5) #checking out the first five observations
```

```python
#Tracking the number of types
typegroup = pokedata.groupby("type1").count()["id"] #Making a count of how many instances each type appears in the type1 column - uses the id column
typegroup #calling it to see if it worked
```

```python
#Making a bar chart for types
fig1 = go.Figure(data=[go.Bar(x = typegroup.index, y = typegroup.values)]) #setting the bar chart to use the index and values from typegroup to create the bar chart
fig1.show()
```

```python
#Creating a color sequence - assigning a color to each type
pokecolors_map = {#Saying to map a color to each type - using hex codes
	'Normal': '#A8A77A',
	'Fire': '#EE8130',
	'Water': '#6390F0',
	'Electric': '#F7D02C',
	'Grass': '#7AC74C',
	'Ice': '#96D9D6',
	'Fighting': '#C22E28',
	'Poison': '#A33EA1',
	'Ground': '#E2BF65',
	'Flying': '#A98FF3',
	'Psychic': '#F95587',
	'Bug': '#A6B91A',
	'Rock': '#B6A136',
	'Ghost': '#735797',
	'Dragon': '#6F35FC',
	'Dark': '#705746',
	'Steel': '#B7B7CE',
	'Fairy': '#D685AD',
}
pokecolors = typegroup.index.map(pokecolors_map) #formally mapping the colors to each type from the typegroup variable - looking at index specifically, not values
pokecolors = list(pokecolors)
pokecolors
```

```python
#Adding color to the Type Bar Chart
fig2 = go.Figure(data=[go.Bar(x=typegroup.index, y=typegroup.values,
 marker_color=pokecolors)]) #setting the colors to align with the pokecolors we set before

fig2.show()
```

```python
#Making a Pie Chart of Pokemon Types

fig3 = go.Figure(
    data=[
    go.Pie(#Saying we want a pie chart
        labels=typegroup.index, #Calling the typegroup index
        values=typegroup.values, #calling the typegroup values
        marker=dict(colors=pokecolors), #Saying we want the colors to be based off pokecolors
        title="Gen 1 Pokemon by Main Type" #Setting the title of the pie chart
        )])

fig3.show()
```

```python
#Checking the null data type
type(pokedata["next_evolution"][150])
```

```python
#Cleaning up the null data

conditions = [#Listing out the conditions for the variable using numpy
    (pokedata['next_evolution'].apply(isinstance, args=(list,))) & (pokedata['prev_evolution'].apply(isinstance, args=(float,))), #Condition 1 = if next evolution is a list and previous evolution is float
    (pokedata['next_evolution'].apply(isinstance, args=(float,))) & (pokedata['prev_evolution'].apply(isinstance, args=(list,))), #Condition 2 = if next evolution is float and previous evoluation is list
    (pokedata['next_evolution'].apply(isinstance, args=(float,))) & (pokedata['prev_evolution'].apply(isinstance, args=(float,))) #Condition 3 = Both next and previous evoluation are floats
]

choices = ["Starter", "Final", "Lone"] #Assigning an order (Start form, Final Form, No Form)
pokedata['evolution_stage'] = np.select(conditions, choices, default="Middle") #Making a column that runs through the conditions and choices laid out earlier - if none of the conditions are met, put Middle
```

```python
#Running the tail to see if it worked
pokedata.tail()
```

```python
#Grouping the Evolution Data
evol = pokedata.groupby("evolution_stage").count()["id"]
evol
```

```python
#Looking at weight type - is a string. Will need to turn into continuous data to work
type(pokedata["weight"][0])
```

```python
#Converting weight into numbers/continuous data
pokedata["weight_kg"] = pokedata["weight"].str.replace(" kg","",regex=False)#Saying to make a new column that is weight, but removes the kg. Regex is saying interpret this as a string of text
pokedata["weight_kg"]=pd.to_numeric(pokedata["weight_kg"]) #Updating the column to be numeric data
```

```python
#Converting height into numbers/continuous data
pokedata['height_m'] = pokedata['height'].str.replace(' m', '', regex=False) #Make a new column for height, minus the m
pokedata['height_m'] = pd.to_numeric(pokedata['height_m']) #Updating the column to be numeric data

pokedata.head() #running to see if it worked
```

```python
#Building a graph using go.Figure - instead of specifying the figure like go.Pie or go.Bar, we put the components in to build the figure for us instead. Allows more freedom

traces = go.Scatter(#Saying we want the traces to take the form of a scatter plot
                x=pokedata["height_m"], #Saying to pull the height data for x
                    y=pokedata["weight_kg"], #Saying to pull the weight data for y
                    mode="markers") #saying to put markers/dots - not lines or text

layout = go.Layout(#Saying we are controlling the layout
                title=dict(text="Pokemon weight and height"), #This is assigning a title 
                   xaxis=dict(title="Weight (in kg)"), #assigning an x-axis label
                   yaxis=dict(title="Height (in m)")) #assigning a y-axis label

fig5 = go.Figure(data=[traces], layout=layout) #We are plugging in what we made earlier - this allowed us to better control the figure content

fig5.show()
```
