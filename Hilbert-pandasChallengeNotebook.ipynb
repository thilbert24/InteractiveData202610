#### Responding to the Murray Reading

##### Question 1 - Write/Render/Evaluate Loop
As discussed in Murray (2017) the write/render/evaluate loop is invaluable to data visualization creators. First, it allows users to speed up the process of working with large datasets by computing and reading data at a much faster pace. Additionally, it allows for quick rendering of visualizations and, when needed, remaking the visualizations in different styles to find the most effective approach. This is partly because of the binary languages of computers. Since computational systems have a limited syntax, it is easy to adjust some values to spit out something different. However, there is a time commitment to learning the syntax that not everyone has. This could limit the re-rendering part, but it would still be faster than doing it by hand.

---

##### Question 2 - Why Dynamic, Interactive Data Visualizations are Useful for Communicating Data
Data visualizations that are both dynamic and interactive are useful for communicating data since they are accessible to the largest population possible. Adhering to Shneiderman's mantra (Murray, 2017), there are some users who are interested in the broad overview of data and do not see value in digging deep into the data. However, by having the interactive elements available to users, the tool becomes valuable to those who need to drill down and see specific data points.

---

##### Question 3 - Murray and Open-Source Tools
I can agree with Murray's sentiment on open-source tools to a point. Open-source tools allow for more customization than proprietary software can, leading to more complex, interactive visualizations. However, I think it really depends on the environment in which they are being used. Approaching this from my experience in higher education, having a lot of "homegrown" tools can be valuable since it can meet a niche need of the organization. However, if only one person has training to operate that tool and leaves, it puts pressure on the unit to hire someone who can continue that work and understand the language(s) used to manage the tool. Because of this, I think proprietary software may be better in some cases since it could be more approachable to less "techie" people.
